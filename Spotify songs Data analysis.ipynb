Data Visualization!

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['font.family'] = 'NanumBarunGothic'
```

```python
plt.figure(figsize=(10, 6))
df1=df.sort_values("Track Duration (ms)",ascending=False)
df1=df1.head(50)
sns.boxplot(data=df1, x='Album Name', y='Track Duration (ms)')
plt.xticks(rotation=90)
plt.title('Distribution of Song Durations Across Albums')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df1=df.sort_values("Track Duration (ms)")
df1=df1.head(50)
sns.boxplot(data=df1, x='Album Name', y='Track Duration (ms)')
plt.xticks(rotation=90)
plt.title('Distribution of Song Durations Across Albums')
plt.show()
```

```python
df['Album Release Date'] = pd.to_datetime(df['Album Release Date'])
df['Year'] = df['Album Release Date'].dt.year
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Year', y='Track Duration (ms)', estimator='mean')
plt.title('Change in Song Durations Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Song Duration (ms)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Popularity'], bins=20, kde=True)
plt.title('Distribution of Popularity Scores')
plt.xlabel('Popularity Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Year', palette='viridis')
plt.title('Trends in Album Release Dates Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Albums Released')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df1=df.sort_values("Popularity",ascending=False)
df1=df1.head(40)
sns.boxplot(data=df1, x='Artist Genres', y='Popularity')
plt.title('Popularity Scores Across Different Genres')
plt.xlabel('Genre')
plt.ylabel('Popularity Score')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Danceability', y='Energy', hue='Popularity', palette='viridis')
plt.title('Correlation Between Danceability and Energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.show()
```

```python
from wordcloud import WordCloud
artist_names = ' '.join(df['Artist Name(s)'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(artist_names)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud of Most Frequent Artist Names')
plt.axis('off')
plt.show()
```

```python
# 8. Distribution of key signatures in the dataset
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Key', palette='Set2')
plt.title('Distribution of Key Signatures')
plt.xlabel('Key')
plt.ylabel('Count')
plt.show()
```

```python
# 10. Distribution of loudness levels in the songs
plt.figure(figsize=(8, 6))
sns.histplot(df['Loudness'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Loudness Levels')
plt.xlabel('Loudness (dB)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 11. Distribution of song tempos (BPM)
plt.figure(figsize=(8, 6))
sns.histplot(df['Tempo'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Song Tempos (BPM)')
plt.xlabel('Tempo (BPM)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 12. Distribution of time signatures
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Time Signature', palette='Set1')
plt.title('Distribution of Time Signatures')
plt.xlabel('Time Signature')
plt.ylabel('Count')
plt.show()
```

```python
# 13. Prevalence of explicit content across genres
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Artist Genres', hue='Explicit', palette='Set2')
plt.title('Prevalence of Explicit Content Across Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Explicit', loc='upper right')
plt.show()
```

```python
# 14. Relationship between song duration and popularity
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Track Duration (ms)', y='Popularity', color='green')
plt.title('Relationship Between Song Duration and Popularity')
plt.xlabel('Song Duration (ms)')
plt.ylabel('Popularity')
plt.show()
```

```python
# 16. Most common artist genres and their popularity
plt.figure(figsize=(12, 6))
top_genres = df['Artist Genres'].value_counts().head(10).index
sns.boxplot(data=df[df['Artist Genres'].isin(top_genres)], x='Artist Genres', y='Popularity', palette='Set3')
plt.title('Top 10 Most Common Artist Genres and Their Popularity')
plt.xlabel('Genre')
plt.ylabel('Popularity')
plt.xticks(rotation=90)
plt.show()
```

```python
# 17. Relationship between speechiness and popularity
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Speechiness', y='Popularity', color='purple')
plt.title('Relationship Between Speechiness and Popularity')
plt.xlabel('Speechiness')
plt.ylabel('Popularity')
plt.show()
```

```python
# 18. Distribution of acousticness across different eras
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Year', y='Acousticness', palette='coolwarm')
plt.title('Distribution of Acousticness Across Different Eras')
plt.xlabel('Year')
plt.ylabel('Acousticness')
plt.xticks(rotation=90)
plt.show()
```

```python
# 19. Trends in album release dates when plotted over time
plt.figure(figsize=(10, 6))
sns.histplot(df['Album Release Date'], bins=20, kde=True)
plt.title('Trends in Album Release Dates Over Time')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# 20. Distribution of track numbers within albums
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Track Number', palette='Pastel1')
plt.title('Distribution of Track Numbers Within Albums')
plt.xlabel('Track Number')
plt.ylabel('Count')
plt.show()
```

```python
# 10. Most common artist genres
plt.figure(figsize=(10, 6))
top_genres = df['Artist Genres'].value_counts().head(10).index
sns.countplot(data=df[df['Artist Genres'].isin(top_genres)], x='Artist Genres', palette='Set3')
plt.title('Top 10 Most Common Artist Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
top_artists = df['Artist Name(s)'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_artists.values, y=top_artists.index, palette='coolwarm')
plt.title('Top Artists by Number of Tracks')
plt.xlabel('Number of Tracks')
plt.ylabel('Artist Name')
plt.show()
```

```python
top_popular_artists = df.groupby('Artist Name(s)')['Popularity'].mean().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_popular_artists.values, y=top_popular_artists.index, palette='Set2')
plt.title('Artists with the Highest Popularity Scores')
plt.xlabel('Average Popularity Score')
plt.ylabel('Artist Name')
plt.show()
```

```python
top_albums = df.groupby('Album Name')['Popularity'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_albums.values, y=top_albums.index, palette='viridis')
plt.title('Top Albums Based on User Engagement (Popularity Sum)')
plt.xlabel('Total Popularity Score')
plt.ylabel('Album Name')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Year', y='Popularity', ci=None, estimator='mean')
plt.title('Trends in Album Popularity Scores Over the Years')
plt.xlabel('Year')
plt.ylabel('Mean Popularity Score')
plt.xticks(rotation=90)
plt.show()
```

```python
top_album_artists = df['Album Artist Name(s)'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_album_artists.values, y=top_album_artists.index, palette='Set3')
plt.title('Top Album Artists by Number of Albums')
plt.xlabel('Number of Albums')
plt.ylabel('Album Artist Name')
plt.show()
```
